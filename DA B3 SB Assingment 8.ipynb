### Qno5:How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

```python
df=pd.read_csv('cars Company.csv')
df
```

```python
df.rename(columns={'Engine HP':'Power','highway MPG':'Effiency'},inplace=True )
```

```python
df = df.groupby('Year')[['Power', 'Effiency']].mean().round()
```

```python
df
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df.index, df['Power'], label='Power', marker='o', color='blue')
plt.plot(df.index, df['Effiency'], label='Effiency', marker='s', color='green')

plt.xlabel('Year')
plt.ylabel('Average Value')
plt.title('Power and Efficiency Over the Years')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
